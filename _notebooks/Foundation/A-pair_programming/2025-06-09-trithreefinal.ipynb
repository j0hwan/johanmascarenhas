  <h2>My Journey Through the Trimesters</h2>

  <h3>Tri&nbsp;1</h3>
  <p>
    I got accustomed to coding platforms and tools and familiarized myself with <code>GitHub</code> for version control.
    Although I was initially slow with projects and learning <code>VS Code</code> as my main IDE,
    by the end of the trimester I could write basic <code>HTML</code> pages, minimal <code>JavaScript</code> functions and <code>Python</code> scripts,
    and use GitHub for cloning, pushing, and committing code.
  </p>

  <h3>Tri&nbsp;2</h3>
  <p>
    I accelerated my workflow and deepened my knowledge of <code>JavaScript</code>, <code>Python</code>, and full-stack development principles.
    I invested significant time in the <em>Travelor</em> project, building a frontend using <code>HTML/CSS</code> and DOM manipulation that connects
    to a backend <code>Flask</code> API. I also learned how to structure routes, handle JSON data, and deploy my work using platforms like <code>Replit</code> and <code>Render</code>.
  </p>

  <h3>Tri&nbsp;3</h3>
  <p>
    For <em>MediPulse</em>, I developed and deployed two production-ready machine learning models:
  </p>
  <ul>
    <li>
      A score-prediction model in <code>scikit-learn</code> that ranks hospitals based on location, safety, experience, and adverse event rates.
      It includes feature weighting, normalization, and logistic regression.
    </li>
    <li>
      A 300-tree <code>RandomForestClassifier</code> that predicts YouTube video performance using title metadata and synonym augmentation
      powered by <code>nltk</code> and <code>WordNet</code>.
    </li>
  </ul>
  <p>
    I also integrated a <code>Gemini Pro</code> API module to generate engagement strategies and created fully responsive frontend
    components in <code>HTML/CSS</code> and <code>Tailwind</code> to interface with each backend service via <code>Flask</code> routes.
  </p>

  <h3>Team Contributions</h3>
  <p>As an <strong>assistant scrum master</strong>, I:</p>
  <ul>
    <li>Shared my ML logic and code snippets, influencing Gavin Copley‚Äôs <code>edit_optimization.py</code> implementation.</li>
    <li>Assisted Daksha Gowda and Thomas Bao in turning backend specs into Figma wireframes and functional UI mockups.</li>
    <li>Led ideation sessions and collected usability feedback to enhance our UI/UX through structured iteration.</li>
    <li>Collaborated with Gavin to manage GitHub Projects, enforce commit discipline, and run stand-ups for sprint accountability.</li>
